```python
# dfs - based


def dfs(st, goal, graph, vis=None):
    if vis is None:
        vis = set()
    
    print(f"Visiting: {st}")
    
    if st == goal:
        return True
    
    vis.add(st)
    
    for child in graph[st]:
        if child not in vis:
            if dfs(child, goal, graph, vis):
                return True
                
    return False

graph1 = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'D'],
    'C': ['A'],
    'D': ['A', 'B']
}

print("Search A to C:")
print("Result:", dfs('A', 'C', graph1))

print("\nSearch A to D:")
print("Result:", dfs('A', 'D', graph1))
```

```python
# dfs - based 



def dfs_v2(st, goal, graph, vis=None):
    if vis is None:
        vis = set()
        
    vis.add(st)
    print(f"Current Node: {st}")
    
    if st == goal:
        return True
        
    for child in graph[st]:
        if child not in vis:
            if dfs_v2(child, goal, graph, vis):
                return True
                
    return False

graph2 = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'E', 'D'],
    'D': ['B', 'C', 'F'],
    'E': ['C', 'F', 'B'],
    'F': ['D', 'E']
}

print("Searching for path from A to F:")
found = dfs_v2('A', 'F', graph2)
print("Path Found:", found)
```
