```python
#Q1
import collections

def bfs(graph, start, goal):
    start = start.upper()
    goal = goal.upper()

    visited = set()
    result = []
    
    queue = collections.deque([start])
    visited.add(start)
    
    while queue:
        
        node = queue.popleft()
        result.append(node)
        
        if node == goal:
            return result
        
        # Explore neighbors
        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
                
    return result

D = {
    "A": ["B", "C", "D"],
    "B": ["A", "E"],
    "C": ["A", "D", "F", "E"],
    "D": ["A", "C", "F"],
    "E": ["B", "C", "A"],
    "F": ["E", "C", "D"]
}

def create_user_graph():
    user_graph = {}
    nodes_count = int(input("Enter number of nodes: "))
    for _ in range(nodes_count):
        node = input("Enter node name: ").upper()
        neighbors = input(f"Enter neighbors for {node} separated by space: ").upper().split()
        user_graph[node] = neighbors
    return user_graph

start_node = input("Enter starting node: ")
goal_node = input("Enter goal node: ")

search_result = bfs(D, start_node, goal_node)

print("\nPath (Nodes visited in BFS order):")
for i in search_result:
    print(i)
```
