```python
import pandas as pd
import numpy as np
d = {
    "name": np.array(["a", "b", "c", "d", "e", "f"]),
    "age": np.array([1, np.nan, 100, 50, 60, 70]),
    "gender": np.array(["m", "f", "none", "m", "f", "none"]),
    "height": np.array([50, 100, 2000, 150, 174, 160]),
    "weight": np.array([50, 100, 200, 80, 54, np.nan])
}
df = pd.DataFrame(d)

print(df.shape)

print(df.size)

df.info()

print(df.describe())

df['weight'] = df['weight'].fillna(df['weight'].mean())

print(df.index.tolist())

filtered_df = df[df['height'] > 100]

print(filtered_df)

result = df.groupby('age')['weight'].agg(['min', 'max', 'mean'])

print(result)
```
