```python
# Given  a 2D numpy array and perform opertaions: range,max,min,sum
import numpy as np

matrix = np.array([[6, 1, 1],
                   [4, -2, 5],
                   [2, 8, 7]])

max_val = np.max(matrix)
min_val = np.min(matrix)
range_val = np.ptp(matrix)
sum_val = np.sum(matrix)
determinant = np.linalg.det(matrix)
inverse = np.linalg.inv(matrix)
eigenvalues = np.linalg.eigvals(matrix)

print(f"Matrix:\n{matrix}")
print(f"Max: {max_val}, Min: {min_val}, Range: {range_val}, Sum: {sum_val}")
print(f"Determinant: {determinant}")
print(f"Inverse:\n{inverse}")
print(f"Eigenvalues: {eigenvalues}")
```
