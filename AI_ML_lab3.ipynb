```python
#Q1
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 

d1 = np.array([1, 2, 3, 4])
d2 = np.array([5, 6, 7, 8])

plt.plot(d1, d2)
plt.bar(d1, d2)
plt.plot(d1, d2, marker='v', color='r', linestyle='dashed')

sns.histplot(x=d1, y=d2) 

plt.show()
```

```python
#Q2
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    'age': [20, 25, 30, 35, 40, 45],
    'height': [150, 160, 165, 170, 175, 180],
    'name': ['A', 'B', 'C', 'D', 'E', 'F']
}
df = pd.DataFrame(data)

sns.scatterplot(x='age', y='height', data=df, hue='name')
plt.title("Seaborn Scatterplot")
plt.show()

sns.displot(x='height', data=df)
plt.title("Height Distribution")
plt.show()

d1 = np.array([10, 20, 30, 40]) 
plt.pie(d1, labels=['Group 1', 'Group 2', 'Group 3', 'Group 4'], autopct='%1.1f%%')
plt.title("Data Distribution Pie Chart")
plt.show()
```

```python
#Q3
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

d = {
    "Customer ID": np.array([1000, 1021, 1022, 1037, 9870]),
    "Customer Name": np.array(['Shiv', 'Piyush', 'Yash', 'Arav', 'Deep']),
    "Product": np.array(['Protein bar', 'Maggi', 'Milk', 'Bread', 'Eggs']),
    "quantity": np.array([2, 5, 6, np.nan, 3]),
    "Cost": np.array([500, 50, 300, 75, 800]),
    "Category": np.array(['Health care', 'grocery', 'dairy', 'grocery', 'grocery'])
}

df = pd.DataFrame(d)
print("Initial DataFrame:")
print(df)

quan = df['quantity'].sum()
cost = df['Cost'].sum()

print(f"\nTotal Quantity: {quan}")
print(f"Total Cost: {cost}")

df['quantity'] = df['quantity'].fillna(df['quantity'].mean())

plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Cost', data=df)
plt.title("Cost by Category")
plt.show()

plt.figure(figsize=(8, 5))
sns.lineplot(x='Category', y='Cost', data=df, marker='o')
plt.title("Trend of Cost across Categories")
plt.show()
```
